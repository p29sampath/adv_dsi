```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from joblib import dump
from src.models.eval_model import eval_model
from sklearn.preprocessing import StandardScaler
%load_ext autoreload
%autoreload 2
```

```python
df=pd.read_csv('./kaggle_assignment/data/raw/train.csv')
```

```python
df = df.drop(['Id_old','Id'],axis=1)
```

```python
df['TARGET_5Yrs'].value_counts()
```

```python
df_minority = df[df['TARGET_5Yrs']==0]
df_majority = df[df['TARGET_5Yrs']==1]
```

```python
[df_minority.shape , df_majority.shape]
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=6669,    # to match majority class
                                 random_state=123) # reproducible results
```

```python
df_minority_upsampled['TARGET_5Yrs'].value_counts()
```

```python
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=1331,     # to match minority class
                                 random_state=123) # reproducible results
```

```python
df_majority_downsampled['TARGET_5Yrs'].value_counts()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
df_upsampled['TARGET_5Yrs'].value_counts()
```

```python
x=df_upsampled.drop('TARGET_5Yrs',axis=1)
y=df_upsampled['TARGET_5Yrs']
```

```python
x_data , x_val ,y_data,  y_val = train_test_split(x, y, test_size=0.2, random_state = 8)
x_train , x_test , y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state = 8)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(random_state=0 )
```

```python
eval_model(lg,x_train,y_train,"Train")
eval_model(lg,x_val,y_val,"Val")
```

```python
lg1 = LogisticRegression(random_state=0,max_iter=1000 )
eval_model(lg1,x_train,y_train,"Train")
eval_model(lg1,x_val,y_val,"Val")
```
